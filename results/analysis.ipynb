```python
build_graph = pandas.read_csv('build_graph1.csv')
#build_graph["name"] = build_graph.dataset.str.cat(build_graph.k.str)

a = build_graph.ix[:,[True, False, True, True]]
# a.plot(kind='bar', x='dataset', title='')

# b = a.groupby(['dataset']).mean()
# b.plot(kind='bar', title='')

fig = plt.figure()
ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
width = 0.4

a = build_graph.ix[:,[True, False, True, False]].groupby(['dataset']).mean()
a.plot(kind='bar', color='red', ax=ax, width=width, title='', position=0)
b = build_graph.ix[:,[True, False, False, True]].groupby(['dataset']).mean()
b.plot(kind='bar', color='blue', ax=ax2, width=width, title='', position=1)
ax.set_ylabel('Time')
ax2.set_ylabel('Layer Size')
```
